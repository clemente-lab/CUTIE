```python
import numpy as np
import pandas as pd
```

```python
meta_map = pd.read_csv('../data/ibd/meta_map.txt', sep = '\t', skiprows = 1)
id_to_meta = pd.Series(meta_map['Exact Match to Standard (* = isomer family)'].values,index=meta_map['Metabolomic Feature']).to_dict()
```

```python
meta_raw = pd.read_csv('../data/ibd/meta_raw.txt', sep = '\t', skiprows=1)
keep_cols = [c for c in meta_raw.columns if c.split('|')[0] == 'PRISM']
meta_raw['# Feature / Sample'] = meta_raw['# Feature / Sample'].map(id_to_meta)
meta_raw = meta_raw.dropna(how='any')
df_meta = meta_raw.set_index('# Feature / Sample')
df_meta = df_meta[keep_cols]
df_meta = df_meta.T
# 390 unique, 468 or so prior
df_meta = df_meta.groupby(lambda x:x, axis=1).sum()
df_meta.head()
df_meta.to_csv('../data/ibd/df_meta.txt', sep = '\t')
```

```python
otu_raw = pd.read_csv('../data/ibd/otu_raw.txt', sep = '\t', skiprows = 1)
keep_cols = [c for c in otu_raw.columns if c.split('|')[0] == 'PRISM']
df_otu = otu_raw.set_index('# Feature / Sample')
df_otu = df_otu[keep_cols]
#nonzero_min = df_otu[ df_otu != 0 ].min()/2
#for col in df_otu.columns.values:
#    df_otu[col] = df_otu[col].replace(0,nonzero_min[col])
#df_otu = np.log(df_otu)
df_otu = df_otu.T
df_otu.head()
df_otu.to_csv('../data/ibd/df_otu.txt', sep = '\t')
```
